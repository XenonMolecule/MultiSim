```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("../../languages_over_time.csv")
df.head()
```

```python
x = df["Year"].to_list()

plt.rc('font', size=16)
plt.rc('axes', titlesize=30)     # fontsize of the axes title
plt.rc('axes', labelsize=16)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=12)    # fontsize of the tick labels
plt.rc('ytick', labelsize=12)    # fontsize of the tick labels
plt.rc('legend', fontsize=12)    # legend fontsize

sns.set_style('darkgrid')

for i,col in enumerate(df):
    if col != "Year":
        if i <= 10:
            plt.plot(x, df[col].to_list(), label =col)
        else:
            plt.plot(x, df[col].to_list(), label =col, linestyle="--")

plt.xlabel("Year")
plt.ylabel("# Text Simplification Papers")
plt.legend()

plt.tight_layout()

plt.savefig("../output/misc/language_over_time.pdf", bbox_inches='tight', pad_inches=0)
```
